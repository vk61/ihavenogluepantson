## Problem: Maximum Product Subarray

Given an integer array `nums`, find a **contiguous non-empty subarray** within the array which has the **largest product**, and return that product.

**Link:**  
https://leetcode.com/problems/maximum-product-subarray/

---

### Example Test Cases

- Input: `nums = [2, 3, -2, 4]` → Output: `6` (subarray `[2, 3]`)  
- Input: `nums = [-2, 0, -1]` → Output: `0`  

```python
from typing import List

class Solution:
    def maxProduct(self, nums: List[int]) -> int:
        res = max(nums)
        currMax = 1
        currMin = 1
        for val in nums:
            if val == 0:
                currMax, currMin = 1, 1
                continue

            tempMax = val * currMax
            tempMin = val * currMin
            currMax = max(tempMax, tempMin, val)
            currMin = min(tempMax, tempMin, val)

            res = max(res,currMax)
        return res
            
```
