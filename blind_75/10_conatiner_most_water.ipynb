## Problem: Container With Most Water

Given an integer array `height` of length `n`, there are `n` vertical lines drawn such that the two endpoints of the _i-th_ line are `(i, 0)` and `(i, height[i])`. Find two lines that together with the x-axis form a container, such that the container contains the **most water**, and return the maximum amount of water a container can store. You may not slant the container.

**Link:**  
https://leetcode.com/problems/container-with-most-water/

---

### Example Test Cases

- Input: `height = [1,8,6,2,5,4,8,3,7]` → Output: `49`  
- Input: `height = [1,1]` → Output: `1`  
- Input: `height = [4,3,2,1,4]` → Output: `16`

```python
from typing import List

class Solution:
    def maxArea(self, height: List[int]) -> int:
        max_area = 0
        max_len = 0
        left,right = 0,len(height)-1
        while left < right:
            area = min(height[left],height[right]) * (right-left)
            max_area = max(max_area,area)

            if height[left] < height[right]:
                left+=1
            else:
                right-=1
        return max_area
            
```

```python
Solution().maxArea([1,8,6,2,5,4,8,3,7])
```

```python
Solution().maxArea([1,1])
```
