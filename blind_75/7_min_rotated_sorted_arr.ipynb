## Problem: Find Minimum in Rotated Sorted Array

Suppose an array of length `n` sorted in ascending order is rotated between 1 and `n` times. Given the rotated array `nums` of **unique** elements, return the **minimum element** of this array. Your solution must run in **O(log n)** time.

**Link:**  
https://leetcode.com/problems/find-minimum-in-rotated-sorted-array/

---

### Example Test Cases

- Input: `nums = [3, 4, 5, 1, 2]` → Output: `1`  
- Input: `nums = [4, 5, 6, 7, 0, 1, 2]` → Output: `0`  
- Input: `nums = [11, 13, 15, 17]` → Output: `11`

```python
from typing import List

class Solution:
    def findMin(self, nums: List[int]) -> int:
        left,right = 0,len(nums)-1
        min_val = nums[0]
        while left <= right:
            if nums[left] < nums[right]:
                min_val = min(min_val,nums[left])
                break
            mid = (left+right) // 2
            min_val = min(min_val,nums[mid])
            if nums[mid] >= nums[left]:
                left = mid+1
            else:
                right = mid-1
        return min_val
```
