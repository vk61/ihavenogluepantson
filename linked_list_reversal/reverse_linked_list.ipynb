```python
# Definition for singly-linked list.
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next
from typing import Optional


class Solution:
    def reverseList(self, head: Optional[ListNode]) -> Optional[ListNode]:
        prev = None
        while head:
            temp = head.next
            head.next = prev
            prev = head
            head = temp
        return prev
```

```python
def createLinkedList(arr):
    head = None
    for val in arr[::-1]:
        if head:
            temp = ListNode(val,next=head)
            head = temp
        else:
            head = ListNode(val)
    return head
def printLinkedList(head):
    curr = head
    print("")
    while curr:
        print(f"{curr.val}","-> "if curr.next else "",end="")
        curr = curr.next
```

```python
head = createLinkedList([1,2,3])
printLinkedList(head)
reversed = Solution().reverseList(head)
printLinkedList(reversed)
```
