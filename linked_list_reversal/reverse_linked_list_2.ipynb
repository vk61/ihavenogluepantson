## Reverse a Sub-list

Given the head of a singly linked list and two integers left and right where left <= right, reverse the nodes of the list from position left to position right, and return the reversed list.

```python


class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next
        
def createLinkedList(arr):
    head = None
    for val in arr[::-1]:
        if head:
            temp = ListNode(val,next=head)
            head = temp
        else:
            head = ListNode(val)
    return head
def printLinkedList(head):
    curr = head
    print("")
    while curr:
        print(f"{curr.val}","-> "if curr.next else "",end="")
        curr = curr.next
```

```python

from typing import Optional

class Solution:
    def reverseBetween(self, head: Optional[ListNode], left: int, right: int) -> Optional[ListNode]:
        curr = head
        leftNode = None
        prev = None
        prevLeftNode = None
        i = 0
        if left==right:
            return head
        while curr:
            i+=1
            temp = curr.next
            if i == left: 
                leftNode = curr
                prevLeftNode = prev
            elif i == right:
                if prevLeftNode:
                    prevLeftNode.next = curr

                leftNode.next = curr.next
                curr.next = prev
                prev = curr
                break
            if leftNode:
                curr.next = prev
            prev = curr
            curr = temp
        if left == 1:
            return prev
        return head
```

```python
head = createLinkedList([1,2,3,4,5])
printLinkedList(head)
reversed = Solution().reverseBetween(head,left=2,right=4)
printLinkedList(reversed)
```
