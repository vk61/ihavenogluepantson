https://leetcode.com/problems/substring-with-concatenation-of-all-words/description/

You are given a string s and an array of strings words. All the strings of words are of the same length.

A concatenated string is a string that exactly contains all the strings of any permutation of words concatenated.

For example, if words = ["ab","cd","ef"], then "abcdef", "abefcd", "cdabef", "cdefab", "efabcd", and "efcdab" are all concatenated strings. "acdbef" is not a concatenated string because it is not the concatenation of any permutation of words.
Return an array of the starting indices of all the concatenated substrings in s. You can return the answer in any order.


It is similar to the anagrams of a string. Instead in the problem, the charcters are words.  the words can be stored in hashmap.

```python
from typing import List
from collections import defaultdict, deque

class Solution:
    def findSubstring(self, s: str, words: List[str]) -> List[int]:
        word_map = defaultdict(int)
        uniq_words = 0
        for word in words:
            if word not in word_map:
                uniq_words+=1
            word_map[word]+=1
        window_start = 0
        word_len = len(words[0])
        
        window_size = len(words)*word_len
        matched = [0]*len(s)
        matched_word_map = []
        window_start_range = len(s)-window_size
        for i in range(window_start_range+1):
            matched_word_map.append(word_map.copy())
        queue_found = deque()
        ans = []
        for window_end in range(0,len(s)):
            word = s[window_end:window_end+word_len]
            # print("w",word,window_start,window_end,matched,uniq_words,window_size,matched_word_map)
            if word in word_map:
                if window_start !=window_end and  window_end <= window_start_range: 
                    queue_found.append(window_end)
                for i in range(window_end,window_start-1,-word_len):
                    # print("i",i)
                    if i <= window_start_range:
                        matched_word_map[i][word]-=1
                        if matched_word_map[i][word]==0:
                            matched[i] +=1
            # print("f",word,window_start,window_end,matched,uniq_words,window_size,word_map)
            if window_end-window_start+1 == window_size:
                window_start_word = s[window_start:window_start+word_len]
                if window_start_word in word_map:
                    if matched[window_start] == uniq_words:
                        ans.append(window_start)
                    matched_word_map[window_start][window_start_word]+=1
                    if matched_word_map[window_start][window_start_word] == 1:
                        matched[window_start] -=1
                if not queue_found:
                    window_start = window_start + 1
                else:
                    window_start = queue_found.popleft()
                
                
                
        return ans
```
